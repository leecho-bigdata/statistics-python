## 3.Chi-square test(적합도)

### 3.1 분할표(contingency table)

```python
observed = god_df['관찰수']
expected = god_df['관찰수'].sum() * god_df['판매비율']
```

```python
np.array(observed)
np.array(expected)
```

```python
stats.chisquare(f_obs=observed, f_exp=expected)
```
