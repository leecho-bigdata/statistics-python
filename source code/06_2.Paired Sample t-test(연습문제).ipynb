## 4.통계분석

### 4.1 차이가 있는 경우(two-sided)

```python
# paired = True : paired sample t-test
pg.ttest(pst_df['ch1al'], pst_df['ch1be'],
         paired = True,
         alternative = "two-sided").round(3)
```

```python
# paired = True : paired sample t-test
pg.ttest(pst_df['ch2al'], pst_df['ch2be'],
         paired = True,
         alternative = "two-sided").round(3)
```

```python
# paired = True : paired sample t-test
pg.ttest(pst_df['ch3al'], pst_df['ch3be'],
         paired = True,
         alternative = "two-sided").round(3)
```

```python
# paired = True : paired sample t-test
pg.ttest(pst_df['ch4al'], pst_df['ch4be'],
         paired = True,
         alternative = "two-sided").round(3)
```
